```python
import pandas as pd


path_data = './../data/tweets.csv'
```

```python
raw_data = pd.read_csv(path_data) 
```

```python
raw_data.head()
```

```python
spliter = lambda line: [item for item in line.split('/') if item and item != 'status']
link_splited = raw_data['link'].map(spliter).to_list()
link_splited_data =  pd.DataFrame(link_splited, columns=['name','link'])
data = raw_data.copy()
data['name'] = link_splited_data['name']
data['link'] = link_splited_data['link']
```
